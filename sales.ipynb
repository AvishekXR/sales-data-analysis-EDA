# Data Exploration and Data Cleaning

```python
# Checking if there is null value in dataset.
sale_data.isna().sum()
```

```python
# Replace full day names with short versions
sale_data['Day Name'] = sale_data['Day Name'].replace({
    'Monday': 'Mon',
    'Tuesday': 'Tue',
    'Wednesday': 'Wed',
    'Thursday': 'Thu',
    'Friday': 'Fri',
    'Saturday': 'Sat',
    'Sunday': 'Sun'
})
```

```python
# Convert Day, Month, Year to string type
sale_data['Day'] = sale_data['Day'].astype(str)
sale_data['Month'] = sale_data['Month'].astype(str)
sale_data['Year'] = sale_data['Year'].astype(str)

# Combine into a single date string in format Day-Month-Year
sale_data['Date_str'] = sale_data['Day'] + '-' + sale_data['Month'] + '-' + sale_data['Year']

# Convert the combined string to datetime
sale_data['Date'] = pd.to_datetime(sale_data['Date_str'], format='%d-%m-%Y')

# Check the results
print(sale_data[['Day', 'Month', 'Year', 'Date']].head())


#Combine Date, Time, and Year into One DateTime Column We merge the separate Date, Time, and Year columns into a single Date column in datetime format.
#This helps simplify time-based analysis and ensures consistency in date-time data.


```

```python
sale_data.drop(columns=['Day', 'Month', 'Year', 'Date_str'], inplace=True)# Dropping Unnecessary columns
```

```python
# calculating total revenue
sale_data['Total_revenue']=sale_data['Units Sold']*sale_data['Price Per Unit']
```

# Bivariate Analysis Questions

**Time vs Sales**

**13.How does revenue change over months**?

```python
# Grouping data by month and summing total revenue
monthly_revenue = sale_data.groupby(sale_data['Date'].dt.to_period('M'))['Total_revenue'].sum().reset_index()
monthly_revenue['Date'] = monthly_revenue['Date'].astype(str)  # Convert Period to string for plotting

# Plotting the revenue trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_revenue, x='Date', y='Total_revenue', marker='o', color='mediumseagreen')
plt.title('📈 Monthly Revenue Trend', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

**14.Are weekend sales higher than weekday sales?**

```python
# Filter weekend and weekday data using 'Day Name' directly
weekend_days = ['Sat', 'Sun']

# Weekend revenue
weekend_revenue = sale_data[sale_data['Day Name'].isin(weekend_days)]['Total_revenue'].sum()

# Weekday revenue
weekday_revenue = sale_data[~sale_data['Day Name'].isin(weekend_days)]['Total_revenue'].sum()

# Create a DataFrame for plotting
day_type_revenue = pd.DataFrame({
    'Day Type': ['Weekday', 'Weekend'],
    'Total Revenue': [weekday_revenue, weekend_revenue]
})
day_type_revenue
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(data=day_type_revenue, x='Day Type', y='Total Revenue', palette='Set2')
plt.title(' Weekend vs Weekday Revenue', fontsize=13, fontweight='bold')
plt.xlabel('Day Type')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```

```python
#We can observe that weekday sales are higher than weekend sales.
```

# Product vs Sales

**16.Is there a relationship between price and units sold?**

```python
correlation = sale_data['Price Per Unit'].corr(sale_data['Units Sold'])
print(f"Correlation between Price Per Unit and Units Sold: {correlation:.2f}")
```

```python
corr_data = sale_data[['Price Per Unit', 'Units Sold', 'Total_revenue']]  
# Calculate correlation matrix
corr_matrix = corr_data.corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

### 📦 Correlation Heatmap: Price Per Unit vs Units Sold

- The heatmap displays the correlation coefficients between key numeric variables.
- It helps to quickly identify whether **Price Per Unit** and **Units Sold** have a positive or negative relationship.
- Values close to +1 or -1 indicate strong relationships, while values near 0 suggest weak or no linear relationship.


**17.Which brands receive the highest average ratings?**

```python
# Calculate average customer rating by brand
avg_rating = sale_data.groupby('Brand')['Customer Ratings'].mean().sort_values(ascending=False).reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_rating, x='Customer Ratings', y='Brand', palette='viridis')
plt.title('📊 Average Customer Ratings by Brand')
plt.xlabel('Average Rating')
plt.ylabel('Brand')
plt.xlim(0, 5)  # Ratings are typically between 0 and 5
plt.show()
```

**20.Which brands/models are most popular in specific cities?**

```python
brand_city_pivot = sale_data.pivot_table(
    index='City',
    columns='Brand',
    values='Units Sold',
    aggfunc='sum',
    fill_value=0
)
brand_city_pivot
```

**21.Do certain cities prefer budget vs premium phones?**

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_city.index, y=avg_price_city.values, palette='coolwarm')
plt.title('Average Price Per Unit by City')
plt.xlabel('City')
plt.ylabel('Average Price (INR)')
plt.xticks(rotation=45)
plt.show()
```
